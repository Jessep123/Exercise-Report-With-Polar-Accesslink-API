```python
import requests
import os
user_id = os.environ.get("POLAR_USER_ID")
access_token = os.environ.get("POLAR_ACCESS_TOKEN")
```

```python
#Getting user information

headers = {
  'Content-Type': 'application/json', 
  'Accept': 'application/json',  
  'Authorization': f'Bearer {access_token}'
}

user_info = requests.get(f'https://www.polaraccesslink.com/v3/users/{user_id}', headers = headers
    ).json()

#Checking out my user data
for key, item in user_info.items():
    print(f'{key} - {item}')
```

```python
exercises_header = {
  'Accept': 'application/json',  'Authorization': f'Bearer {access_token}'
}
exercise_data = requests.get(f'https://www.polaraccesslink.com/v3/exercises', headers = exercises_header).json()

for item in exercise_data:
    print(item['id'])
```
